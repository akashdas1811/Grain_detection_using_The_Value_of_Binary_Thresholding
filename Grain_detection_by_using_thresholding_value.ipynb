```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('./inout images/image1.jpg')
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

blurred = cv2.GaussianBlur(gray, (3, 3), 0) #Gaussian Blur to smooth out the image
thresh = cv2.threshold(blurred, 60, 255, cv2.THRESH_BINARY_INV)[1]
```

```python
plt.imshow(thresh, cmap= 'gray')
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contour_img = cv2.drawContours(thresh, contours, -1, (255, 255, 255), 2) #change the colour tuple (255, 255, 255) to change contour colour
plt.imshow(contour_img, cmap='gray')
```

```python
inv_img = cv2.bitwise_not(thresh)
plt.imshow(inv_img, cmap = 'gray') #change the threshold in cv2.threshold to change results in the image
```
